```python
import matplotlib.pyplot as plt

# Function to add bar values on top of bars
def add_bar_values(ax):
    for p in ax.patches:
        ax.annotate('{:.2f}'.format(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Data for Linear Regression
lr_train_mse = 255.649916
lr_test_mse = 257.438812
lr_train_r2 = 0.670877
lr_test_r2 = 0.667375

# Data for Decision Tree
dt_train_mse = 0.001033
dt_test_mse = 24.806184
dt_train_r2 = 0.999999
dt_test_r2 = 0.967949

# Plotting MSE
plt.figure(figsize=(12, 6))

# Plot for Linear Regression
ax1 = plt.subplot(1, 2, 1)
lr_mse_bars = ax1.bar(['Train', 'Test'], [lr_train_mse, lr_test_mse], color=['blue', 'orange'])
plt.title('Linear Regression - Mean Squared Error')
plt.ylabel('MSE')
add_bar_values(ax1)

# Plot for Decision Tree
ax2 = plt.subplot(1, 2, 2)
dt_mse_bars = ax2.bar(['Train', 'Test'], [dt_train_mse, dt_test_mse], color=['blue', 'orange'])
plt.title('Decision Tree - Mean Squared Error')
plt.ylabel('MSE')
add_bar_values(ax2)

plt.tight_layout()
plt.show()

# Plotting R-squared
plt.figure(figsize=(12, 6))

# Plot for Linear Regression
ax3 = plt.subplot(1, 2, 1)
lr_r2_bars = ax3.bar(['Train', 'Test'], [lr_train_r2, lr_test_r2], color=['green', 'red'])
plt.title('Linear Regression - R-squared')
plt.ylabel('R-squared')
add_bar_values(ax3)

# Plot for Decision Tree
ax4 = plt.subplot(1, 2, 2)
dt_r2_bars = ax4.bar(['Train', 'Test'], [dt_train_r2, dt_test_r2], color=['green', 'red'])
plt.title('Decision Tree - R-squared')
plt.ylabel('R-squared')
add_bar_values(ax4)

plt.tight_layout()
plt.show()
```
